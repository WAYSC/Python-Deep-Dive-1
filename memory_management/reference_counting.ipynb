```python
myvar = 10
print(id(myvar))  # 输出类似 1718348528（十进制地址）
print(hex(id(myvar)))  # 输出类似 0x666bef0（十六进制地址）
```

总结一下：

- 引用计数是 Python 跟踪内存对象被引用次数的机制；
- 计数 + 1：新增引用（比如变量赋值）；
- 计数 - 1：引用消失（比如变量重新赋值）；
- 计数 = 0：内存自动回收；
- 查看计数：`sys.getrefcount()` 要减 1，`ctypes` 方法更准确。

```python
# 第一步：创建对象10，myvar引用它，引用计数=1
myvar = 10
# 第二步：othervar引用myvar指向的地址，引用计数=2
othervar = myvar
```

```python
# othervar不再引用10的地址，引用计数=1
othervar = None
# myvar不再引用10的地址，引用计数=0
myvar = None
```

```python
import sys
a = [1, 2, 3]
# 注意：调用getrefcount时会临时创建一个引用，所以结果要减1
print(sys.getrefcount(a))  # 输出2，实际引用计数=1
```

```python
import ctypes

def ref_count(address):
    return ctypes.c_long.from_address(address).value

a = [1, 2, 3]
a_id = id(a)
print(ref_count(a_id))  # 输出1，准确反映当前引用计数
```

```python
import ctypes

def ref_count(address):
    return ctypes.c_long.from_address(address).value

# 创建对象，引用计数=1
a = [1, 2, 3]
a_id = id(a)
print(ref_count(a_id))  # 输出1

# 新增引用，引用计数=2
b = a
print(ref_count(a_id))  # 输出2

# 新增引用，引用计数=3
c = a
print(ref_count(a_id))  # 输出3

# c不再引用，引用计数=2
c = 10
print(ref_count(a_id))  # 输出2

# b不再引用，引用计数=1
b = None
print(ref_count(a_id))  # 输出1

# a不再引用，引用计数=0
a = None
print(ref_count(a_id)) # 输出1

a = None
print(ref_count(a_id)) # 输出0
```
