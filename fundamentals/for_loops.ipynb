# Part1: for循环的基本语法

```python
for i in [1, 2, 3, 4]:
    print(i)
```

# Part2：for循环 vs while循环

while：只要满足条件就一直循环（不知道次数）

for：把一堆东西，挨个过一遍（知道要循环多少次）

# Part3：遍历字符串

```python
for char in "hello":
    print(char)
```

# Part4：for+range()

## 4.1 rang(n) -> 从0到n-1

```python
for i in range(5):
    print(i)
```

## 4.2 range(start, stop)

```python
for i in range(2, 6):
    print(i)
```

## 4.3 range带步长

```python
for i in range(0, 10, 2):
    print(i)
```

## range总结：range(start, stop, step)

start：起始数字（默认0，可以省略）

stop：结束数字（必须写，并且不包含本身）

step：步长（每次跳几步，默认1，可以为负数）

```python
for i in range(10, 0, -2):
    print(i)
```

# Part5：用for循环计算总和

```python
total = 0
for i in [1, 2, 3, 4]:
    total += i
print(total)
```

# Part6：找到列表的最大值

```python
highest = 0
numbers = [1, 3, 5, 2, 4]
for n in numbers:
    if n > highest:
        highest = n
print(highest)
```

# Part7：统计字符出现次数

```python
count = 0
for char in "banana":
    if char == "a":
        count += 1
print(count)
```

# Part8：循环变量名可以随便取

```python
for x in [1, 2, 3]:
    print(x)

for number in [1, 2, 3]:
    print(number)

for 苹果 in [1, 2, 3]:
    print(苹果)
```
