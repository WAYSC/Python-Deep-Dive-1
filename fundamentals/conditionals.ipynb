# if-else语句

## 语法结构

```python
'''
if 条件表达式:
    # 条件为True时执行的代码块
    语句
else:
    # 条件为False时执行的代码块
    语句
'''
```

## 实战例子：判断数字大小

```python
a = 3
if a < 5:
    print("a小于5")
else:
    print("a大于等于5")
```

## 进阶用法：嵌套 if 语句（多层判断）

##  实战例子：判断成绩等级

假设要求：

- 分数＜60：不及格
- 60≤分数＜80：及格
- 分数≥80：优秀

```python
score = 75
if score < 60:
    print("成绩：不及格")
else:
    # 进入else说明score≥60，继续判断
    if score < 80:
        print("成绩：及格")
    else:
        print("成绩：优秀")
```

## 高效优化：elif 语句（多选一逻辑）

```python
'''
if 条件1:
    执行代码1
elif 条件2:
    执行代码2
elif 条件3:
    执行代码3
else:
    # 所有条件都不满足时执行
    执行代码4
'''
```

## 实战例子：分段判断数字范围

```python
a = 12
if a < 5:
    print("a大于等于0，小于5")
elif a < 10:
    print("a大于等于5，小于10")
elif a < 15:
    print("a大于等于10，小于15")
elif a < 20:
    print("a大于等于15，小于20")
else:
    print("a大于等于20")
```

#### 优势对比：

| 实现方式 | 代码复杂度     | 可读性 | 适用场景           |
| -------- | -------------- | ------ | ------------------ |
| 嵌套 if  | 高（多层缩进） | 差     | 条件存在依赖关系   |
| elif     | 低（平级结构） | 好     | 条件互斥（多选一） |

## 简洁技巧：三元表达式（一行实现 if-else）

```python
'''变量 = 表达式1 if 条件 else 表达式2'''
```

## 实战例子：简化判断逻辑

```python
a = 4
if a < 5:
    b = "a小于5"
else:
    b = "a大于等于5"
print(b)  # 输出：a小于5
```

```python
a = 4
b = "a小于5" if a < 5 else "a大于等于5"
print(b)  # 输出：a小于5
```

仅适用于 "二选一赋值" 的简单逻辑，复杂代码（如多句执行语句）不建议使用，否则会降低可读性。
