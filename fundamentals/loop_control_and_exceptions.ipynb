# break, continue, try

## break

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

## continue

```python
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
for i in range(3):
    for j in range(5):
        if j == 2:
            break
        print(f"i={i}, j={j}")
```

## try

```python
'''
try:
    # 可能出错的代码
excepte:
    # 出错后执行的代码
finally:
    # 无论是否出错都会执行
'''
```

```python
try:
    print(10 / 0)
except:
    print("发生错误：不能除以零")
```

```python
try:
    x = int("abc")
except:
    print("转换失败")
finally:
    print("无论是否出错，我都会执行")
```

## try和continue搭配的例子

```python
numbers = ["10", "20", "abc", "40"]

for n in numbers:
    try:
        value = int(n)
    except:
        print(f"跳过无法转换的值：{n}")
        continue
    print(f"成功转换：{value}")
```

## try和break搭配的例子

```python
data = ["5", "6", "e", "7"]

for d in data:
    try:
        num = int(d)
    except:
        print("发生错误，中断循环")
        break
    print("正常处理：", num)
```
