```python
x = 1 # 初始值：给x一个起点
while x <= 5: # 循环条件
    print(x)
    x += 1 # 条件的更新
```

# 实战案例

## 案例1：用户输入验证（最常用场景）

```python
num = int(input("请输入1-5之间的数字："))
while num < 1 or num > 5:
    num = int(input("请输入1-5之间的数字："))
print("你输入了：",num)
```

## 案例2：指令监听（知道输入指定值才退出）

```python
command = ""
while command != "Q":
    command = input("输入指令（输入Q退出）：")
print("程序已退出")
```

## 案例3：无限循环+break

```python
while True:
    text = input("输入内容（输入stop结束）：")
    if text == "stop":
        break
```

## 案例4：倒计时功能（条件驱动的固定循环）

```python
count = 5
while count > 0:
    print("倒计时：", count)
    count -= 1
print("发射！")
```

适用场景：

1、循环次数不确定（比如用户输入验证、指令监听）

2、循环依赖条件变化（比如倒计时）

3、需要持续验证用户输入

4、不断监听事件

5、想要实现无限循环（搭配break）
