# 1. 验证不可变对象（字符串）传参

```python
# 定义处理字符串的函数
def process(s):
    # 打印函数内 s 的初始内存地址
    print("函数内 s 初始内存地址：", id(s))
    # 尝试拼接字符串，修改 s
    s = s + " world"
    # 打印函数内 s 的最终内存地址
    print("函数内 s 最终内存地址：", id(s))
    # 打印函数内 s 的最终值
    print("函数内 s 最终值：", s)

# 主程序定义字符串变量
myvar = "Hello"
# 打印主程序 myvar 的内存地址
print("主程序 myvar 初始内存地址：", id(myvar))
# 调用函数
process(myvar)
# 打印主程序 myvar 的最终值
print("主程序 myvar 最终值：", myvar)
```

# 2. 验证可变对象（列表）传参

```python
# 定义处理列表的函数
def modify_list(lst):
    # 打印函数内 lst 的初始内存地址
    print("函数内 lst 初始内存地址：", id(lst))
    # 原地追加元素，修改列表
    lst.append(100)
    # 打印函数内 lst 的最终内存地址
    print("函数内 lst 最终内存地址：", id(lst))
    # 打印函数内 lst 的最终值
    print("函数内 lst 最终值：", lst)

# 主程序定义列表变量
mylist = [1, 2, 3]
# 打印主程序 mylist 的内存地址
print("主程序 mylist 初始内存地址：", id(mylist))
# 调用函数
modify_list(mylist)
# 打印主程序 mylist 的最终值
print("主程序 mylist 最终值：", mylist)
```

# 3. 验证不可变容器（元组）中的可变元素

```python
# 定义处理元组的函数
def modify_tuple(t):
    # 打印函数内 t 的初始内存地址
    print("函数内 t 初始内存地址：", id(t))
    # 修改元组第一个元素（列表）的内容
    t[0].append(100)
    # 打印函数内 t 的最终内存地址
    print("函数内 t 最终内存地址：", id(t))
    # 打印函数内 t 的最终值
    print("函数内 t 最终值：", t)

# 主程序定义元组（第一个元素是列表，第二个是字符串）
mytuple = ([1, 2], "a")
# 打印主程序 mytuple 的内存地址
print("主程序 mytuple 初始内存地址：", id(mytuple))
# 调用函数
modify_tuple(mytuple)
# 打印主程序 mytuple 的最终值
print("主程序 mytuple 最终值：", mytuple)
```

# 总结

1. 不可变对象（字符串、整数、元组等）：函数传参时，传递的是引用，函数内 “修改” 本质是创建新对象，不会影响原始变量，无意外副作用，更安全；
2. 可变对象（列表、字典等）：函数传参时，传递的是引用，函数内的原地修改会直接改变原始变量，易产生副作用，使用时需谨慎；
3. 特殊情况：不可变容器（如元组）的结构不可变，但内部的可变元素（如列表）内容可以被修改，开发中需格外注意区分。
