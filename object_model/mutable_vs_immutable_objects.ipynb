```python
# 无需指定变量类型，myVar指向字符串对象"Hello"
myVar = "Hello"
# 查看当前变量引用对象的类型：<class 'str'>
print(type(myVar))
```

```python
# 合法：将myVar的引用改为整数对象10
myVar = 10
# 查看当前变量引用对象的类型：<class 'int'>
print(type(myVar))
```

```python
# 合法：将myVar的引用改为复数对象
myVar = 3 + 4j
# 查看当前变量引用对象的类型：<class 'complex'>
print(type(myVar))
```

```python
# 合法：将myVar的引用改为函数对象（lambda表达式）
myVar = lambda x: x ** 2
# 查看当前变量引用对象的类型：<class 'function'>
print(type(myVar))
# 调用函数，输出：4
print(myVar(2))
```

# 1. 重赋值的本质：创建新对象，更换引用

```python
# 步骤1：创建整数对象10，变量a指向该对象
a = 10
# 打印变量a引用对象的内存地址（每次运行可能不同，仅作参考）
print(f"a=10时，对象内存地址：{id(a)}")
# 打印变量类型：<class 'int'>
print(f"a=10时，对象类型：{type(a)}")
```

```python
# 步骤2：重赋值a=15，创建新整数对象15，变量a更换引用指向新对象
a = 15
# 打印内存地址：与之前不同，说明指向了新对象
print(f"a=15时，对象内存地址：{id(a)}")
# 打印变量类型：<class 'int'>
print(f"a=15时，对象类型：{type(a)}")
```

```python
# 步骤3：重赋值a = a + 1，先计算右侧表达式，创建新对象16，再更换引用
a = a + 1
# 打印内存地址：再次变化，指向新的整数对象16
print(f"a=a+1时，对象内存地址：{id(a)}")
# 打印变量值：16
print(f"a=a+1时，变量值：{a}")
```

# 三、 对象的可变性：可变对象 vs 不可变对象

- **可变对象（Mutable）**：对象创建后，其内部状态（数据内容）可以被修改，且修改后对象的内存地址保持不变。
- **不可变对象（Immutable）**：对象创建后，其内部状态（数据内容）无法被修改，若要改变 “值”，只能创建新的对象，更换变量引用。

不可变对象：无法修改内部状态，只能创建新对象

```python
# 创建字符串对象"Hello"，变量s指向该对象
s = "Hello"
# 打印内存地址
print(f"s='Hello'时，内存地址：{id(s)}")
```

```python
# 看似修改字符串，实则创建新的字符串对象"Hello World"
s = s + " World"
# 打印内存地址：与之前不同，指向新对象
print(f"s='Hello World'时，内存地址：{id(s)}")
```

```python
# 尝试通过索引修改字符串，直接报错（TypeError）
s[0] = 'h'  # 非法操作，不可变对象无法修改内部内容
```

```python
# 创建元组对象(1, 2, 3)，变量t指向该对象
t = (1, 2, 3)
# 打印内存地址
print(f"t=(1,2,3)时，内存地址：{id(t)}")
```

```python
# 尝试通过索引修改元组元素，直接报错（TypeError）
t[0] = 10  # 非法操作，元组不可变
```

```python
# 看似修改元组，实则创建新的元组对象
t = t + (4, 5)
# 打印内存地址：与之前不同，指向新对象
print(f"t=(1,2,3,4,5)时，内存地址：{id(t)}")
```

可变对象：修改内部状态，内存地址不变

```python
# 创建列表对象[1, 2, 3]，变量my_list指向该对象
my_list = [1, 2, 3]
# 打印内存地址
print(f"初始列表内存地址：{id(my_list)}")
# 打印列表内容：[1, 2, 3]
print(f"初始列表内容：{my_list}")
```

```python
# 使用append()方法修改列表内部内容，添加元素4
my_list.append(4)
# 打印内存地址：与初始地址一致，说明还是同一个对象
print(f"append后列表内存地址：{id(my_list)}")
# 打印列表内容：[1, 2, 3, 4]
print(f"append后列表内容：{my_list}")
```

```python
# 注意：使用+号拼接列表，会创建新对象
my_list2 = [1, 2, 3]
print(f"my_list2初始内存地址：{id(my_list2)}")
# 拼接列表，创建新对象
my_list2 = my_list2 + [4]
print(f"拼接后my_list2内存地址：{id(my_list2)}")
print(f"拼接后my_list2内容：{my_list2}")
```

```python
# 创建字典对象，变量my_dict指向该对象
my_dict = {"key1": 1, "key2": "a"}
# 打印内存地址
print(f"初始字典内存地址：{id(my_dict)}")
# 打印字典内容：{'key1': 1, 'key2': 'a'}
print(f"初始字典内容：{my_dict}")
```

```python
# 添加新的键值对，修改字典内部状态
my_dict["key3"] = 3.14
# 打印内存地址：与初始地址一致
print(f"添加键值对后字典内存地址：{id(my_dict)}")
# 打印字典内容：{'key1': 1, 'key2': 'a', 'key3': 3.14}
print(f"添加键值对后字典内容：{my_dict}")
```

```python
# 修改已有键的值，修改字典内部状态
my_dict["key1"] = 100
# 打印内存地址：仍与初始地址一致
print(f"修改值后字典内存地址：{id(my_dict)}")
# 打印字典内容：{'key1': 100, 'key2': 'a', 'key3': 3.14}
print(f"修改值后字典内容：{my_dict}")
```

易错点：不可变容器中可以包含可变元素

```python
# 创建元组t，内部包含两个列表（可变对象）
t = ([1, 2], [3, 4])
# 打印元组初始内容：([1, 2], [3, 4])
print(f"元组初始内容：{t}")
# 打印元组内存地址
print(f"元组初始内存地址：{id(t)}")
```

```python
# 无法修改元组的元素（无法替换列表对象）
t[0] = [1, 2, 3]  # 非法操作，报错TypeError
```

```python
# 可以修改元组内部可变对象（列表）的状态
t[0].append(3)  # 给第一个列表添加元素3
t[1].append(5)  # 给第二个列表添加元素5
```

```python
# 打印元组修改后内容：([1, 2, 3], [3, 4, 5])
print(f"元组修改后内容：{t}")
# 打印元组内存地址：与初始地址一致，说明元组对象未改变
print(f"元组修改后内存地址：{id(t)}")
```

# 总结

1. **类型系统**：Python 是动态类型语言，类型绑定对象而非变量；静态类型语言（Java）类型绑定变量，灵活性低但安全性高。
2. **变量重赋值**：本质是创建新对象，更换变量的引用，而非修改原有对象，小整数（[-5,256]）会被缓存复用。
3. **对象可变性**：不可变对象（int、str、tuple）无法修改内部状态，可变对象（list、set、dict）可修改内部状态且地址不变。
4. **易错点**：不可变容器（tuple）可包含可变元素，容器自身结构不可变，但内部可变元素的状态可修改。
