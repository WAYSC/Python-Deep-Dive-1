# Part5：给类添加属性！用init方法初始化

```python
class Person:
    # self 必须写在第一个参数，代表实例本身
    def __init__(self, name, age):
        # 给实例绑定“name”属性
        self.name = name
        # 给实例绑定“age”属性
        self.age = age
```

怎么使用？在创建实例时直接传入参数！

```python
p = Person("Alice", 20) # 相当于一个叫“Alice”，年龄是20的人

# 访问实例的属（用实例.属性的格式）
print("名字：", p.name)
print("年龄：", p.age)
```

# Part6：给类添加方法！让实例动起来

```python
class Person:
    def __init__(self, name):
        self.name = name
    
    # 实例方法：第一个参数必须是self
    def greet(self):
        print(f"你好呀！我叫{self.name}，很高兴认识你！")
```

```python
p1 = Person("Bob")
p2 = Person("Lisa")

# 让实例执行greet方法
p1.greet()
p2.greet()
```

# Part7：划重点！类和实例的区别

```python
class Person:
    def __init__(self, name):
        self.name = name

p1 = Person("Bob")
p2 = Person("Lisa")

# 给p1单独添加一个age属性
p1.age = 30

print("p1的年龄：", p1.age)
print("p2的年龄：", p2.age)
```

实例属性只属于该实例，不会影响其他实例！
