# 验证Python对小整数的缓存（-5~256）

```python
a = 10
b = 10
```

```python
print("a的地址：", id(a))
print("b的地址：", id(b))
```

```python
print("a和b的内存地址是否相同：", a is b)
```

```python
print("a和b的值是否相等：", a == b)
```

# Python对超出-5~256的大整数是不会缓存的
